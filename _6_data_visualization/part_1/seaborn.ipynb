```python
import seaborn as sns;
import matplotlib.pyplot as plt; 
```

```python
sns.set_theme(); # first work 
```

```python
sns.get_dataset_names(); # all data in seaborn 
```

```python
tips = sns.load_dataset("tips");
print(type(tips));
print(tips.head())
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker", # ye alag se highlight kar dega smokers ko in color 
    size="size",
    style="smoker" # ye smkers ke style  ko badal dega
);
```

```python
 # line plot 
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="line",
    hue="smoker", # ye alag se highlight kar dega smokers ko in color 
    size="size",
    style="smoker" # ye smkers ke style  ko badal dega
);
```

```python
# relational plots -------------> scatter and line plots 
tips = sns.load_dataset("tips");

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
#line plots 
flights = sns.load_dataset("flights");
flights.head();
sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
);
```

```python
#bar plot 
tips.head()
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex" 
)
```

```python
#box plot 
sns.boxplot(
     data=tips,
    x="day",
    y="tip",
)
```

```python
# histograms 

penguins = sns.load_dataset("penguins");
sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
    
)
```

```python
#heat map 
flight_pivot=flights.pivot(index="month",columns="year",values="passengers");
print(flight_pivot);
sns.heatmap(
    data=flight_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"
    
)
```

```python
#modern seaborn 
fig,ax = plt.subplots()
sns.lineplot(
    data=sns.load_dataset("tips"),
    x="day",
    y="total_bill",
    hue="sex",
    ax=ax,
    marker="o",
    errorbar=None # ci ko hide krne ke liye
)
ax.set_title("BoxPlot: total bill by sex");
ax.set_xlabel("Day of the week");
ax.set_ylabel("Total Bill ($)");
plt.legend(title="sex");

plt.show()
```

```python
#modern seaborn 
fig,ax = plt.subplots()
sns.boxplot(
    data=sns.load_dataset("tips"),
    x="day",
    y="total_bill",
    hue="sex",
    ax=ax,
   
)
ax.set_title("BoxPlot: total bill by sex");
ax.set_xlabel("Day of the week");
ax.set_ylabel("Total Bill ($)");
plt.legend(title="sex");

plt.show()
```
