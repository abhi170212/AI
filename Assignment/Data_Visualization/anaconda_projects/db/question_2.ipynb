```python
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
pen = sns.load_dataset("penguins")
pen.head()
```

```python
sns.lineplot(data=pen,x="flipper_length_mm",y="body_mass_g",hue="species")
```

```python
sns.scatterplot(
    data=pen,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species"
)

plt.title("Flipper Length vs Body Mass")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Body Mass (g)")
```
