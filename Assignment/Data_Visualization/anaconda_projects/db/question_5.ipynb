```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tip = sns.load_dataset("tips")
```

```python
tip.head()
```

```python
matrix = tip.pivot_table(
    index="day",
    columns="time",
    values="tip",
    aggfunc="mean"
)

sns.heatmap(matrix, annot=True)
plt.show()
```
