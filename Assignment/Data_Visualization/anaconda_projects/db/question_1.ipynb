```python
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
normal = np.random.normal(loc=70, scale=10, size=200)

plt.hist(
    normal,
    bins=10,
    color='green',
    edgecolor='black',
    alpha=0.7
)

val = np.mean(normal)

plt.axvline(
    x=val,
    color='red',
    linestyle='--',
    linewidth=2,
    label=f"Mean = {val:.2f}"
)

plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Histogram of Normally Distributed Scores")
plt.legend()

plt.show()
```
