```python
import pandas as pd;
import numpy as np;
```

```python
s=pd.Series([1,3,23,6,10])
print(s)
print(f"The value at index 2 is {s[2]}")
print(s.index)
```

```python
named_series = pd.Series([19,10,30,17,15],index=["adam","eve","charlie","bob","messi"])
print(named_series["messi"])

print(named_series.values)
print(named_series.index)
```

```python
#properties of series 
ser=pd.Series([1,4,5])

print(ser+10)
print(ser*10)

print(ser[ser>2])
```

```python
#----------------------------------------------------------------------------------------------
#datframes 
info = {
    "Name":["Adam","Eve","Bob"],
    "Marks":[90,98,91],
    "Grades":['B','A+','A']
}
df = pd.DataFrame(info)
print(df.columns)
print(df.index)

#df using lists and arrays 
arr= np.array([[10,20,30],[40,50,60],[70,80,90]])
df2 = pd.DataFrame(arr,columns=["m1","m2","m3"])
df2

lst = [[200,300,400],[500,600,700]]
df3= pd.DataFrame(lst,columns=["col1","col2","col3"])
df3
```
